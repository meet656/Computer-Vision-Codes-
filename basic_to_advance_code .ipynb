## SET :- 1 (Q_No:-i):-Display the relevant images in different color modes

```python
def lib_fuc():
    import cv2
    from matplotlib import pyplot as plt
    import numpy as np
    global cv2, plt, np

def filter_fun():
    lib_fuc()
    global img
    img = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg")

    
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

   
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 3, 1)
    plt.title("RGB")
    plt.imshow(rgb)
    plt.axis("off")

    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plt.subplot(2, 3, 2)
    plt.title("Gray")
    plt.imshow(gray, cmap="gray")
    plt.axis("off")

    _, bw = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    plt.subplot(2, 3, 3)
    plt.title("Black & White")
    plt.imshow(bw, cmap="gray")
    plt.axis("off")
 
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    plt.subplot(2, 3, 4)
    plt.title("HSV (shown)")
    plt.imshow(hsv_rgb)
    plt.axis("off")


    plt.tight_layout()
    plt.show()

filter_fun()
```

## SET :- 1 (Q_No:-ii):-perform image smoothning using mean , median and gaussian blurr methods(coding and library)

```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg")

# Convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply smoothing
mean = cv2.blur(img_rgb, (5,5))
median = cv2.medianBlur(img_rgb, 5)
gaussian = cv2.GaussianBlur(img_rgb, (5,5), 0)
# Display
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Mean Blur")
plt.imshow(mean)
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Median Blur")
plt.imshow(median)
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Gaussian Blur")
plt.imshow(gaussian)
plt.axis('off')

plt.show()
```

## Set :-2  (Q_No:-ii):-perform contrast stretching, gray level slicing 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg",0)

# Contrast Stretching
min_val = np.min(img)
max_val = np.max(img)

stretch = ((img - min_val) / (max_val - min_val)) * 255
stretch = np.uint8(stretch)

# Show
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(stretch, cmap='gray')
plt.title("Stretched")

plt.show()
```

## Set :-3  (Q_No:-ii):-Detect edges using gradient operators (Laplacian  and canny using library as well as using convolution with their masks)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg", 0)

lap = cv2.Laplacian(img, cv2.CV_64F)#Laplacian
canny = cv2.Canny(img,100,200)#canny

kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
conv = cv2.filter2D(img, -1, kernel)#masks

plt.subplot(1,3,1)
plt.imshow(lap, cmap='gray')
plt.title("Laplacian")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(canny, cmap='gray')
plt.title("Canny")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(conv, cmap='gray')
plt.title("Mask")
plt.axis("off")

plt.show()
```

## Set :-4 (Q_No:-ii):-Perform simple, Otsu thresholding using coding as well as using libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg", 0)

# Simple
ret, th = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

# Otsu
ret, otsu = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Manual
manual = np.zeros_like(img)
manual[img > 127] = 255

plt.subplot(1,3,1)
plt.imshow(th, cmap='gray')
plt.title("Simple")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(otsu, cmap='gray')
plt.title("Otsu");plt.axis("off")


plt.subplot(1,3,3)
plt.imshow(manual, cmap='gray')
plt.title("Manual")
plt.axis("off")

plt.show()
```

## Set :-5 (Q_No:-ii):-Perform  Image addition and subtraction 

```python
import cv2
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg")
img2 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images (1).jpg")

# Check images
if img1 is None or img2 is None:
    print("Error: Image not found")
else:
    # Resize img2 same as img1
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    # Convert to RGB
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

    # Addition
    add = cv2.add(img1, img2)
    sub = cv2.subtract(img1, img2)

    # Display
    plt.figure(figsize=(8,5))

    plt.subplot(2,4,1)
    plt.imshow(img1)
    plt.title("Image 1")
    plt.axis("off")

    plt.subplot(2,4,2)
    plt.imshow(img2)
    plt.title("Image 2")
    plt.axis("off")

    plt.subplot(2,4,3)
    plt.imshow(add)
    plt.title("Addition")
    plt.axis("off")

    plt.subplot(2,4,4)
    plt.imshow(sub)
    plt.title("Subtraction ")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg")
img2 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images (1).jpg")

if img1 is not None and img2 is not None:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))#important 

    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

    add = cv2.add(img1, img2)
    sub = cv2.subtract(img1, img2)

    titles = ["Image 1", "Image 2", "Add", "Sub"]
    images = [img1, img2, add, sub]

    for i in range(4):
        plt.subplot(2,2,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis("off")

    plt.tight_layout()
    plt.show()
else:
    print("Error: Image not found")
```

## Set :-5 (Q_No:-ii):-Perform  bitwise operations on image  

```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images.jpg")
img2 = cv2.imread(r"C:\Users\Meet\OneDrive\Desktop\images (1).jpg")

# Resize (same size important)
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Convert to RGB
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Bitwise operations
and_img = cv2.bitwise_and(img1, img2)
or_img  = cv2.bitwise_or(img1, img2)
xor_img = cv2.bitwise_xor(img1, img2)
not_img = cv2.bitwise_not(img1)

# Display (short loop)
titles = ["AND", "OR", "XOR", "NOT"]
images = [and_img, or_img, xor_img, not_img]

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
